```python
import scipy.stats as stats
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pylab
```

```python
pd.options.display.float_format = '{:.10f}'.format
data = pd.read_csv('30min_database.csv')
car = data.drop('etiqueta',axis=1)
etiquetas = data['etiqueta']
```

```python
data
```

```python
car.skew(axis = 0, skipna = True)
```

```python
def normality(input, caracteristica):
    plt.figure(figsize=(10,5))
    plt.subplot(1,2,1)
    sns.kdeplot(input[caracteristica])
    plt.subplot(1,2,2)
    stats.probplot(input[caracteristica],plot=pylab)
    plt.savefig()
    plt.show()
```

```python
for feat in car.columns:
  normality(car,feat)
```

```python
for feat in car.columns:
  normality(car,feat)
```

```python
car.columns
```

```python
boxcox_columns = ['rr_mean', 'sdrr', 'sdsd', 'rmssd', 'dam', 'vlf', 'lf', 'hf', 'lfhf', 'sd1', 'sd2', 'sd12', 'rec', 'lmean', 'lmax', 'entropia',
       'captura', 'determinismo']
```

```python
for feat in boxcox_columns:
  car[feat+'_boxcox'],parameters=stats.boxcox(car[feat])
  normality(car,feat+'_boxcox')
```

```python
car['nn50_sqrt'] = np.sqrt(data['nn50'])
car['pnn50_sqrt'] = np.sqrt(data['pnn50'])
normality(car,'nn50_sqrt')
normality(car,'pnn50_sqrt')
```

```python
car
```

```python
car_transform = car.drop(['rr_mean', 'sdrr', 'sdsd', 'rmssd', 'nn50', 'pnn50', 'dam', 'vlf', 'lf',
       'hf', 'lfhf', 'sd1', 'sd2', 'sd12', 'rec', 'lmean', 'lmax', 'entropia',
       'captura', 'determinismo'], axis=1)
```
